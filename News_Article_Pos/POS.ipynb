```python
import spacy
from collections import Counter
nlp = spacy.load('en_core_web_sm')
```

```python
file = open('news.txt')
```

```python
doc = file.read()
doc = nlp(doc)
```

```python
print(doc)
```

```python
print("WORDS"+"               " + "WORDS POS"+ "           " + "WORD TAG")
for token in doc:
    print(f"{token.text:{20}}{token.pos_:{20}}{token.tag_:{20}}")
```

```python
words = [token.text for token in doc if not token.is_stop and not token.is_punct]
word_c = Counter(words)
words_count = word_c.most_common(500)
print (words_count)
```
